## Hangman Implementation

```python
from agentquest.benchmarks.hangman import HangmanDriver, HangmanUtils

word = HangmanUtils.load_data(category="6 letters")[0]
driver = HangmanDriver(goal=word)
print("The goal word is: ", word)
```

```python
# Initiating the game with reset method call
obs = driver.reset()
print(obs.output)

while not obs.success and obs.can_proceed:
    human_input = input()
    # obs = driver.step(HangmanAction(value=human_input))
    obs = driver.step_raw(human_input)
    print(obs.output)
```

```python
driver.metrics.export(
    repetition_function_kwargs={"num_execution_steps": 10, "theta_a": 1}
)
```
